```python
from langchain.embeddings import OpenAIEmbeddings
import pandas as pd
```

```python
embeddings = OpenAIEmbeddings()
```

```python
greetings_df = pd.read_csv('/Users/aadya/Documents/GitHub/FAQChatbot/src/data/Greetings.csv')
```

```python
data_df = pd.read_csv('/Users/aadya/Documents/GitHub/FAQChatbot/src/data/GST FAQs 2.csv')
data_df
```

```python
import nltk
from nltk.stem.lancaster import LancasterStemmer
def cleanup(sentence):
        word_tok = nltk.word_tokenize(sentence)
        stemmed_words = [LancasterStemmer().stem(w) for w in word_tok]
        return ' '.join(stemmed_words)
```

```python
combined_df = pd.concat([greetings_df,data_df], ignore_index=True)  
combined_df      
```

```python
combined_df['clean_question'] = combined_df['Question'].apply(lambda x : cleanup(x))
question_list = combined_df['clean_question'].to_list()
len(question_list)
```

```python
doc_result = embeddings.embed_documents(question_list)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

query = 'It is really great to interact you'
query_result = embeddings.embed_query(query)
```

```python
sims = cosine_similarity(doc_result, [query_result])
sims
```

```python
import numpy as np
ind = np.argmax(sims)
print(ind)
combined_df['Answer'][ind]
```
